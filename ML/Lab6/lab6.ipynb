### Задание 1

Докажем, что задача логистической регресси является выпуклой задачей машинного обучения. Для этого вспомним, что задача машинного обучения $(H,Z,l)$ называется выпуклой, если множество гипотез $H$ является выпуклым, и $l$ является выпуклой для любого $z \in Z$.

Тройка $(H,Z,l)$ для логистической регрессии представляется следующим образом:

$$
H = R^d, \quad Z = X \times Y = {R}^{d+1} \quad l(w, (x, y)) = ln(1 + e^{-y<w,x>})
$$

Очевидно, множество гипотез $H = R^d$ выпукло.

Функция потерь в логистической регрессии выглядит следующим образом:

$$
l(w, (x,y)) =  ln(1 + e^{- y<w, x>})
$$

Рассмотрим функцию:

$$
g(x) = ln(1+e^{-yx})
$$

Первая производная функции:

$$
g'(x) = -\frac{y}{e^{yx}+1}
$$

Вторая производная функции:

$$
g''(x) = \frac{y^{2}e^{yx}}{(e^{yx}+1)^2} \geq 0
$$

По критерию выпуклости $g''(x) \geq 0 \Rightarrow g(x)$ - выпуклая.
По лемме о композиции выпуклой и линейной функции, функция $l(w, (x,y)) =  ln(1 + e^{- y<w, x>})$ - выпуклая, так как $g(x)=ln(1+e^{-yx})$ - выпуклая, а $<w,x>$ - линейная.

Таким образом, задача логистической регрессии - выпуклая.

### Задание 3

1) Пусть функция $f(x): C \rightarrow R$ (где $C$ - выпуклое множество) выпукла, т.е.

$$
\forall u,v \in C : \quad f (\alpha u + (1 - \alpha)v) \leq \alpha f (u) + (1 - \alpha)f (v), \forall \alpha \in [0,1].
$$

Рассмотрим надграфик функции $f(x)$ - множество, которое представляется в следующем виде:

$$
epi \ f = \{(x,y) \in C \times R \ \ | \ \ y \geq f(x)\}
$$

Докажем, что надграфик $f(x)$ будет выпуклым множеством. Для этого нам надо показать, что:

$$
\forall (x_1, y_1), (x_2, y_2) \in epi \ f : \quad (\alpha x_1 + (1 - \alpha)x_2, \alpha y_1 + (1 - \alpha)y_2) \in epi \ f, \forall \alpha \in [0,1].
$$

Действительно:

$$
\alpha y_1 + (1 - \alpha)y_2 \geq \alpha f(x_1) + (1 - \alpha)f(x_2) \geq f(\alpha x_1 + (1 - \alpha)x_2) \Rightarrow (\alpha x_1 + (1 - \alpha)x_2, \alpha y_1 + (1 - \alpha)y_2) \in epi \ f 
$$

2) Пусть надграфик функции $f(x)$ является выпуклым множеством. Докажем, что $f(x)$ - выпуклая функция. От противного:

Допустим, что:

$$
\exists u,v \in C : \quad f (\alpha u + (1 - \alpha)v) > \alpha f (u) + (1 - \alpha)f (v), \forall \alpha \in (0,1).
$$

Так как точки $(u, f(u)), (v, f(v))$ принадлежат надграфику $f(x)$, то должно выполняться:

$$
f (\alpha u + (1 - \alpha)v) \leq \alpha f (u) + (1 - \alpha)f (v), \forall \alpha \in (0,1).
$$

Противоречие. Значит, $f(x)$ - выпуклая функция.
