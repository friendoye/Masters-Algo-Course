Рассмотрим произвольный элемент из $H_c$ $y^* = \{y_1, y_2, ..., y_d\}, y_i \in \{-1,+1\}$. Тогда гипотеза $$h^* = sign(<w,x>+b) = sign(\sum((1,1,...,1) * x_k) - 0.5) \in H$$ где $x_k$ - элемент, у которого $y_k = +1$. 2) Покажем, что любое $C$ размера $d + 1$ нельзя разукрасить. От противного: предположим, что $C$ можно разукрасить. $C = \{x_1, x_2, ..., x_d, x_{d+1}\}$.
