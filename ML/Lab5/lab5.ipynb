### Задание 1

В алгоритме k-fold параметр k влияет на качество выбора подходящей гипотезы и на оценку $|L_D - L_V|$.

Большой k лучше выбирать, когда есть много априорных знаний и мы больше заинтересованы в уменьшении estimation error.

Маленький k лучше выбирать, когда есть мало априорных знаний и мы хотим более точно знать оценку $L_D$ полученных гипотез (т.е. какой алгоритм лучше подойдет).

### Задание 2

На тренировочной выборке обучаются наши алгоритмы, на валидационной мы можем увидеть, какие из них мы должны 
выбрать или в каких из них мы должны изменить параметры, чтобы наш $L_V$ оказался минимальным. На тестовой выборке мы уже оцениваем $L_D$.

Тестовая выборка нужна, чтобы увидеть насколько хорошо подходит данная модель для решения нашей задачи.

### Задание 3
Преимущество leave-all-out состоит в том, что:
  * данный метод не зависит от того, как мы разбивает нашу выборку на k частей
  * усредняет $L_D$ по всевозможным способам разбиения выборки, в то время как k-fold усредняет только по k возможным валидационным выборкам. Это дает нам возможность лучше понять, когда алгоритм склонен к переобучению, а когда - нет.

Недостатком leave-all-out является то, что он пригоден только для маленьких выборок, так как он будет прогонять алгоритм обучения $2^m$ раз, где m - размер выборки. Также является недостатком то, что в подсчете итоговой $L_D$ учавствуют слишком маленькие тренировочные и валидационные выборки, которые толком не говорят ничего о качествах алгоритма обучения (стоит отметить, что у них будут маленькие веса в итоговорм подсчете, так что это недостаток не так критичен).
