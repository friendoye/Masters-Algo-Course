Напишем KnnClasifier вместе с методом кросс-валидации, причем зададим возможность классификатору нормализовать данные с помощью параметра ***normalize***.

```python
import random
import numpy as np
from collections import Counter

def calc_risk(classifier, features):
    if (len(features) == 0):
        raise "Zero features for calc_risk!"
    summa = 0.0
    for f in features:
        summa += (classifier.predict(f[0]) != f[1])
    return summa / len(features)

def kFold(classifier, features, k):
    np.random.shuffle(features)
    splits = np.array_split(features, k)
    train_risk = []
    validation_risk = []
    for t in range(k):
        validation_split = splits[t]
        train_split = np.concatenate(np.delete(splits, t))
        classifier.train(train_split)
        validation_risk.append(calc_risk(classifier, validation_split))
    avg_train_risk = sum(train_risk) / k
    avg_validation_risk = sum(validation_risk) / k
    return avg_validation_risk

def eucl_distance(x, y):
    return sum([(x0 - y0) ** 2 for (x0, y0) in zip(x, y)]) ** 0.5

class KnnClassifier:

    def __init__(self, k, normalize=False):
        self.train_model = []
        self.k = k
        self.normalize = normalize
    
    def train(self, features):
        self.train_model = np.array(features.copy())
        if self.normalize:
            arr = np.array(list(self.train_model[:, 0]))
            self.min_features_bound = arr.min(0)
            self.features_ptp = arr.ptp(0)
            self.train_model[:, 0] = list((arr - arr.min(0)) / arr.ptp(0))

    def predict(self, x):
        if self.normalize:
            x = (x - self.min_features_bound) / self.features_ptp
        nearest_features = sorted(self.train_model, key=lambda record: eucl_distance(record[0], x))
        counter = Counter([record[1] for record in nearest_features[:self.k]])
        #print "Predicting %s with %s" % (x, nearest_features[:self.k])
        #print "Result: %s" % counter.most_common(1)[0][0]
        return counter.most_common(1)[0][0]
```

Также напишем функцию, которая будет прогонять данные из папки *data* и строить график ошибок для k-NN.

```python
from matplotlib import pyplot as plt

def file_line_to_feature_with_label(line):
    literals = line.split(",")
    return [np.array([float(x) for x in literals[:-1]]), literals[-1]]

def examine_kNN_errors(k_for_kfold, k_NN_max, normalize=False):
    train_file_path = "data/train.data"
    test_file_path = "data/test.data"
    features = []
    test_features = []
    with open(train_file_path) as train_file:
        features = [file_line_to_feature_with_label(s.strip()) for s in train_file.readlines()]
    with open(test_file_path) as test_file:
        test_features = [file_line_to_feature_with_label(s.strip()) for s in test_file.readlines()]

    k_fold_errors = []
    test_errors = []
    x = range(1, k_NN_max + 1)
    for k_for_knn in x:
        classifier = KnnClassifier(k_for_knn, normalize)
        k_fold_errors.append(kFold(classifier, features, k_for_kfold))
        classifier.train(np.array(features))
        test_errors.append(calc_risk(classifier, test_features))
        #print("K-Fold Error for k-NN(%s): %s" % (k_for_knn, k_fold_errors[-1]))
        #print("Test Error for k-NN(%s): %s" % (k_for_knn, test_errors[-1]))

    min_cv_error = min(k_fold_errors)
    optimal_k = k_fold_errors.index(min_cv_error)
    print("Optimal K: %s" % optimal_k)
    print("Min CV error: %s" % min_cv_error)
    print("Test error for Min CV error: %s" % test_errors[optimal_k - 1])

    plt.plot(x, k_fold_errors, label="Cross validation error")
    plt.plot(x, test_errors, label="Test error")
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., title="k-NN errors")
    plt.show()
```

```python
examine_kNN_errors(k_for_kfold=10, k_NN_max=100, normalize=False)
```

Понятие "Проклятие размерности" говорит о том, что с ростом размерности пространства для того, чтобы хорошо восстановить признаки исследуемых объектов по выборке (найти какие-то общие признаки, закономерности,  зависимости), количество объектов должно расти ***эскпоненциально***. Это значит, например, что если для k-NN в многомерном пространстве у нас не будет действительно очень большой выборки, то наш классификатор "переобучится". Поэтому всегда лучше выбирать наиболее полезные признаки объектов и для них строить классификатор (Даже плохой классификатор будет показывать хорошие результаты на хорошей выборке, нежели самый лучший классификатор из возможный на плохой). Также, при росте размерности, "шум" на выборке будет очень сильно влиять на ошибочность классификатора.

Для борьбы с этим "проклятием" можно:
  * Выбрать "лучшие" признаки и строить классификатор по ним
  * Посмотреть, лежат ли объекты $n$-мерного пространства на некоторой гиперплоскости или достаточно близко от нее. Затем перевести эти объекты в $n-1$-мерное пространство, с учетом расположения данной гиперплоскости
  * Использовать *Multidimensional scaling method*, т.е. подход, который переводит выборку в пространства меньшей размерности так, чтобы отношения расстояний между объектами оставались примерно такими же, как и в оригинальной выборке (расстояния между всеми объектами изменились одинаково примерно в $\epsilon$ раз).
